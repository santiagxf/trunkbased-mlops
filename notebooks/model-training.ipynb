## Running the training routine

```python
from jobtools.arguments import file2namespace
from hatedetection.train.trainer import train_and_evaluate
```

```python
train_params = file2namespace('../jobs/hatedetection/train.params.yml')
```

```python
import mlflow
```

```python
mlflow.set_experiment("hate-pt-speech-troubleshouting")
```

```python
with mlflow.start_run():
    results = train_and_evaluate(input_dataset='../datasets/portuguese-hate-speech-tweets/sample/2019-05-28_portuguese_hate_speech_hierarchical_classification.csv',
                                 eval_dataset='../datasets/portuguese-hate-speech-tweets-eval/sample/*.csv',
                                 params=train_params)
```
